##### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns 
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
tips
```

```python
tips.to_csv('tips.csv')
```

```python
tips.head()
```

```python
# create a scatter plot 
import matplotlib.pyplot as plt 

sns.scatterplot(x='total_bill',y='tip',data=tips,color='purple')
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
## Categorical Plots
## BAr Plot
sns.barplot(x='day',y='total_bill',data=tips,color='orange')
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
## Box Plot
sns.boxplot(x="time",y='total_bill',data=tips,color='magenta')
```

```python
## Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips,color='gold')
```

```python
### Histograms
sns.histplot(tips['total_bill'],bins=5,kde=True,color='violet')
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=False)
```

```python
sns.kdeplot(tips['tip'],fill=True,color='lightblue')
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
tips
```

```python
# Heatmap 
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True)
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
