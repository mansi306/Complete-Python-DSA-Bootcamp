#### Pandas - DataFrame and series 

Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. 
It provides two primary data structures: 
1) Series : A Series is a one-dimensional array-like object
2) DataFrame : DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd 
```

```python
## Series
# A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.
data = [1,2,3,4,5,6]
series = pd.Series(data, index = ['A','B','C','D','E','F'])
print("Series \n",series)
print(type(series))
```

```python
# create a series from Dictionary 
data = {'a':1,'b':2,'c':3}
series_data = pd.Series(data)
print(series_data)
```

```python
## Dataframe
## create a Dataframe from a dictionary oof list
data={
    'Name':['John','jocky','Jack'],
    'Age':[25,30,45],
    'City':['Bangalore','New York','Florida']
}

df = pd.DataFrame(data)
# print(df)
print(type(df))
df
```

```python
## Create a Data frame From a List of Dictionaries

data=[
    {'Name':'Krish','Age':32,'City':'Bangalore'},
    {'Name':'John','Age':34,'City':'Hydrabad'},
    {'Name':'Bappy','Age':32,'City':'Pune'},
    {'Name':'JAck','Age':32,'City':'Delhi'}]

df = pd.DataFrame(data)
print(type(df))
df
```
